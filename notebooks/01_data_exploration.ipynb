```python
# استيراد المكتبات
import pandas as pd
import matplotlib.pyplot as plt

# تحميل البيانات
data = pd.read_csv('../data/Language Detection.csv')

# عرض أول 5 صفوف
print("أول 5 صفوف من البيانات:")
print(data.head())

# عرض معلومات عن البيانات
print("\nمعلومات عن البيانات:")
print(data.info())

# عرض توزيع اللغات
print("\nتوزيع اللغات:")
print(data['Language'].value_counts())

# رسم توزيع اللغات
data['Language'].value_counts().plot(kind='bar')
plt.title("توزيع اللغات في البيانات")
plt.xticks(rotation=45)
plt.show()
```
